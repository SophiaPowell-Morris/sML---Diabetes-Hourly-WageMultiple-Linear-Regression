```python
#Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

#Importing the dataset
data = pd.read_csv("diabetes.csv")
data.head(10)
```

## Identify the IV's and DV

```python
X = data.iloc[:,[0,1,2,3,4,5,6,7]].values
y = data.iloc[:,[8]].values 
```

## Training and Test Samples

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2)

print ("Training data:",X_train.shape, y_train.shape)
print ("Test data:",X_test.shape, y_test.shape)
```

```python
X_train
```

## Applying Scalers to the training and test sample sets

```python
# importing scaling modules
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

### 1.  MinMaxScaler

```python
# Creating an object of the MinMaxScalar function
scaler = MinMaxScaler()

# Fitting and transforming scaler fucntion on the training dataset
X_train = scaler.fit_transform((X_train))

minmax_df_new = pd.DataFrame(X_train, columns = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 
                                                   'BMI', 'DiabetesPedigreeFunction', 'Age'])
minmax_df_new
```

```python
# Transforming the test dataset using the training dataset fit 
X_test = scaler.transform(X_test)
X_test_minmax_new = pd.DataFrame(X_test, columns = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 
                                                   'BMI', 'DiabetesPedigreeFunction', 'Age'])
X_test_minmax_new
```

### 2. StandardScaler

```python
scaler2 = StandardScaler()
X_train = scaler2.fit_transform((X_train))
standard_df_new = pd.DataFrame(X_train, columns = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age'])
standard_df_new
```

```python
X_test = scaler2.transform(X_test)
X_test_std_new = pd.DataFrame(X_test, columns = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 
                                                   'BMI', 'DiabetesPedigreeFunction', 'Age'])
X_test_std_new
```

## Creating the Multiple Linear Regression Model

```python
X_train = X_train.reshape(-1, 8)
y_train = y_train.reshape(-1, 1) 

# create an empty linear regression model like below and give it a good variable name
diabetes_model = LinearRegression()

# to create the model, we use fit(X,y)
model = diabetes_model.fit(X_train,y_train)

print('Intercept: \n', model.intercept_)
print('Coefficients: \n', model.coef_)
```

### Modelled linear equation for dependent variable prediction

* **outcome** = 0.34 + 0.066(Pregnancies) + 0.18(Glucose) - 0.046(BloodPressure) - 0.0002(SkinThickness) - 0.037(Insulin) + 0.1(BMI) + 0.049(DPF) + 0.029(Age)

### Predictions for the Test set

```python
predictions = diabetes_model.predict(X_test)
```

```python
predictions[0:5]
```

```python
predictions[predictions < 0.5] = 0
predictions[predictions >= 0.5] = 1
```

```python
predictions[0:5]
```

## Accuracy Check of the Predictive Model

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(y_test, predictions)
acc
```

The model is achieving an predictive accuracy of 81%.
